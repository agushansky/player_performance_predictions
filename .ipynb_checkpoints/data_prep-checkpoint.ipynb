```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, PowerTransformer
```

```python
cwd = os.getcwd()
# load data
fpath = "./data/raw/1_Combined_Player_Per90_Cleaned.csv"
if os.path.isfile(fpath) == False:
    print('Error - fpath is invalid')
else:
    df = pd.read_csv(fpath)
```

```python
# one hot encode league and prev_season_league

league = pd.get_dummies(df.League, prefix='League')
prev_league = pd.get_dummies(df.prev_season_league, prefix='Prev_League')
df = pd.concat([df, league, prev_league], axis=1)
df.drop(columns=["League", "prev_season_league"], inplace=True)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.drop(columns=["Born"], inplace=True)
```

```python
# remember to add team play style cluster
exclude_cols = ["PlayerID", "Season", "Player", "Nation", "Pos", "Squad"]
```

```python
df = PowerTransformer.fit_transform(df[-exclude_cols])
```

```python
df = PolynomialFeatures(degree=2).fit_transform(df)
```
